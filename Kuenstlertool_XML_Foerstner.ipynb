```python
from bs4 import BeautifulSoup
```

```python
image_collection = BeautifulSoup(open("kueShortExport.xml"), 'lxml')
```

```python
for entry in image_collection.find_all("h1:document"):
    print(entry)
    print("\n----------------------------\n")
```

```python
for entry in image_collection.find_all("h1:document"):
    for field in entry.find_all("h1:field"):
        print(field)
        print(field.attrs)
        print("The type:", field.attrs["type"])
        print("The value", field.attrs["value"])
        print("\n------End of field-------\n")
    print("\n ============== End of document entry ==============\n")
```
